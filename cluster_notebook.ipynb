```python
import pandas as pd

path = r'C:\\Users\shuweilin\Desktop\cluster4windows\data\for_cluster_with_label.csv'
df = pd.read_csv(path, header=0)

print(df.shape[0])
total = df.shape[0]
print('--'*5)
df = df.groupby('Group_ID')
groups = df.groups.keys()

for group_idx in groups:
    temp_df = df.get_group(group_idx)
    print('Group_'+str(group_idx))
    print(100*temp_df.shape[0]/total)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = r'C:\\Users\shuweilin\Desktop\cluster4windows\data\group_center.csv'
center_df = pd.read_csv(path, header=0)
fig=plt.figure(figsize=(7,4.5))
plt.ylim([0,2500])

for idx in range(center_df.shape[0]):
    plt.plot(np.arange(96)/4, center_df.iloc[idx, 1:-1].values, lw=2, label='Group %d' % (idx+1))

plt.xlabel('Period', size=14)
plt.ylabel('Power (w)', size=14)
fig.suptitle('Group center', fontsize=16)
plt.title('K = 3', size=14)
plt.legend(loc='upper center', bbox_to_anchor=(0.6,0.95),ncol=3,fancybox=True,shadow=True)
plt.show()
```

```python
from base64 import b64encode
import json

ENCODING = 'utf-8'
IMAGE_NAME, JSON_NAME = r'C:\Users\shuweilin\Center.svg', r'C:\Users\shuweilin\result.json'
with open(IMAGE_NAME, 'rb') as jpg_file:
    byte_content = jpg_file.read()
    
base64_bytes = base64.b64encode(byte_content)
base64_string = base64_bytes.decode(ENCODING)

raw_data = {}
raw_data["name"] = IMAGE_NAME
raw_data["image_base64_string"] = base64_string
json_data = json.dumps(raw_data, indent=2)

with open(JSON_NAME, 'w') as json_file:
    json_file.write(json_data)with open(JSON_NAME, "r") as json_file:
    raw_data = json.load(json_file)
    
image_base64_string = raw_data["image_base64_string"]
image_data = base64.b64decode(image_base64_string)

with open(r'C:\Users\shuweilin\result.svg', 'wb') as jpg_file:
    jpg_file.write(image_data)
```
